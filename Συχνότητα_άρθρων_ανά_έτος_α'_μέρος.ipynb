```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/CSVFILES/filtered_ergatikaparapolitika.csv')
```

```python
df["datetime"] = pd.to_datetime(df["date"] + " " + df["time"], errors="coerce")

# Εξαγωγή έτους
df["year"] = df["datetime"].dt.year
```

```python
year_counts = df["year"].value_counts().sort_index()

# Γράφημα
plt.figure(figsize=(8, 4))
plt.bar(year_counts.index, year_counts.values, color="darkblue")
plt.title("Αριθμός άρθρων ανά έτος - Παραπολιτικά")
plt.xlabel("Έτος")
plt.ylabel("Πλήθος άρθρων")
plt.xticks(year_counts.index, rotation=45)
plt.tight_layout()
plt.show()
```
