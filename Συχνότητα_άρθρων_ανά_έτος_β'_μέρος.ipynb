```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
ktdf = pd.read_csv('/content/drive/MyDrive/CSVFILES/filtered_ergatikakathimerini.csv')
```

```python
ktdf["datetime"] = pd.to_datetime(ktdf["date"] + " " + ktdf["time"], errors="coerce")

# Εξαγωγή έτους
ktdf["year"] = ktdf["datetime"].dt.year
```

```python
year_counts = ktdf["year"].value_counts().sort_index()

# Γράφημα
plt.figure(figsize=(8, 4))
plt.bar(year_counts.index, year_counts.values, color="yellow")
plt.title("Αριθμός άρθρων ανά έτος - Kαθημερινή")
plt.xlabel("Έτος")
plt.ylabel("Πλήθος άρθρων")
plt.xticks(year_counts.index, rotation=45)
plt.tight_layout()
plt.show()
```
