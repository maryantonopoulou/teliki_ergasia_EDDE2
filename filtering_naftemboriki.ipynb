```python
df=pd.read_csv('/content/drive/MyDrive/CSVFILES/naftemporiki_articles.csv')
```

```python
# Μετατροπή ημερομηνίας
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Ορισμός χρονικού διαστήματος (π.χ. 2016–2023)
start_date = "2015-01-01"
end_date = "2025-06-05"
df = df[(df["date"] >= start_date) & (df["date"] <= end_date)]

# Συνένωση τίτλου + κειμένου για αναζήτηση λέξεων
df["full_text"] = df["title"].astype(str) + " " + df["text_body"].astype(str)

# Keywords για filtering
positive_keywords = [
    "εργάτης", "εργαζόμενος", "εργοτάξιο", "εν ώρα εργασίας", "εργασία",
    "κατάστημα", "delivery", "τεχνικός", "εργατικό ατύχημα","εργατικό δυστύχημα" "εργασία"
]

negative_keywords = [
    "τροχαίο", "σπίτι",
    "οικιακό","τροχαίο ατύχημα", "celebrities" , "celebrity" , "αθλητισμός"
]

def is_work_related(text):
    text = str(text).lower()
    has_positive = any(kw in text for kw in positive_keywords)
    has_negative = any(kw in text for kw in negative_keywords)
    return has_positive and not has_negative

# Εφαρμογή φίλτρου
df["is_ergatiko"] = df["full_text"].apply(is_work_related)

# Τελικό φιλτράρισμα
df_filtered = df[df["is_ergatiko"] == True]

# Αποθήκευση
df_filtered.to_csv("filtered_ergatika_clean.csv", index=False)

print(f"Τελικός αριθμός άρθρων: {len(df_filtered)} από σύνολο {len(df)}")
```
