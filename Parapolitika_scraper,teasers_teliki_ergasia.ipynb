```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
```

```python
def get_webpage_soup(url):

  headers = {
        "User-Agent": "Mozilla/5.0"
    }
  response = requests.get(url, headers=headers)

  if response.status_code != 200:
    print(f"Error: {response.status_code}")
    return None
  else:
    print(f"{url} was scraped successfully")
    soup = BeautifulSoup(response.text, 'html.parser')
    return BeautifulSoup(response.text, 'html.parser')
```

```python
def get_teaser_elements_from_articlesL(articlesL):
  articles_dataL = []
  for article in articlesL:
    try:
      title = article.find("h3").text.strip()
    except:
      title=""
    try:
      article_url = article.find("h3").a["href"]
      if "https" not in article_url:
        article_url = f"https://www.parapolitika.gr{article_url}"
    except:
      article_url=""
    try:
      datetime_str = article.find("time").time["datetime"]
      date_str = datetime_str.split("T")[0]
      time_str = datetime_str.split("T")[1].split("+")[0]
    except:
      date_str = ""
      time_str = ""
    try:
      author = article.find("span", {"class": "article__author"}).text.strip()
    except:
      author = ""
    try:
      section = article.find("a", class_=lambda c: c and "article__category" in c).text.strip()
    except:
      section = ""
    try:
        teaser = article.find("div", class_="article__summary").text.strip()
    except:
        teaser = ""
    articleD = {"title": title, "article_url": article_url, "date": date_str, "time": time_str, "author": author, "section": section,"teaser":teaser}
    articles_dataL.append(articleD)

  return articles_dataL
```

```python
def scrape_parapolitika_search_pages(search_term, start_page_nbr, end_page_nbr):
  all_articles_dataL = []

  print(f"\rScraping started", end="")

  for page_nbr in range(start_page_nbr, end_page_nbr+1):
    url = f"https://www.parapolitika.gr/apotelesmata-anazitisis/?q={search_term}&pg={page_nbr}"

    print(f"\rScraping page {page_nbr}: {url}" + " "*2, end="")

    soup = get_webpage_soup(url)

    articlesL = soup.find("div", {"class": "row articlesGrid articlesGrid--triple"}).find_all("article")

    page_articles_dataL = get_teaser_elements_from_articlesL(articlesL)

    all_articles_dataL.extend(page_articles_dataL)

    delay = random.uniform(1, 3)
    time.sleep(delay)
  df = pd.DataFrame(all_articles_dataL)
  print(f"\rScraping completed. Fetshed {len(df)} articles.")

  return df
```

```python
ergatika_atiximatadf=scrape_parapolitika_search_pages('εργατικό+ατύχημα',1,171)
ergatika_atiximatadf
```
